```python
import numpy as np
import matplotlib.pyplot as plt

# Création d'une grille avec motif "U"
def creer_grille_u(n):
    grille = np.zeros((n, n), dtype=int)
    centre = n // 2
    grille[centre - 1, centre - 1:centre + 2] = 1
    grille[centre:centre + 2, centre - 1] = 1
    grille[centre:centre + 2, centre + 1] = 1
    return grille

# Affichage
def afficher_grille(grille):
    plt.figure(figsize=(6, 6))
    plt.imshow(grille, cmap='gray', origin='lower')
    plt.axis('off')
    plt.show()
```

```python
# Calcul des voisins vivants
def matrice_voisins(grille):
    total = np.zeros_like(grille)
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx != 0 or dy != 0:
                total += np.roll(np.roll(grille, dx, axis=0), dy, axis=1)
    return total

# Règles d'évolution
def evolution(grille):
    voisins = matrice_voisins(grille)
    return ((grille == 1) & ((voisins == 2) | (voisins == 3)) | ((grille == 0) & (voisins == 3))).astype(int)

# Afficher k étapes
def jouer_k_iterations(grille, k, afficher=False):
    for _ in range(k):
        grille = evolution(grille)
        if afficher:
            afficher_grille(grille)
    return grille
```

```python
# Tester le jeu avec affichage
taille = 30  # Taille de la grille
grille = creer_grille_u(taille)

# Affichage de l'état initial
afficher_grille(grille)

# Faire évoluer la grille pendant 10 étapes avec affichage
grille = jouer_k_iterations(grille, 10, afficher=True)
```
